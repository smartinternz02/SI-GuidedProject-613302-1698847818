## Check for Null Values

```python
#Checking for Null values
df.isnull().sum()
```

```python
#replacing null values
df['Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Married']=df['Married'].fillna(df['Married'].mode()[0])
df['Dependents']=df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].mode()[0])
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0])
df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].mode()[0])
```

## Outlier Detection

```python
#Outlier Detection
df.describe()
```

```python
sns.boxplot(df.ApplicantIncome)
```

```python
q1 = df.ApplicantIncome.quantile(0.25)
q3 = df.ApplicantIncome.quantile(0.75)
IQR = q3-q1
upper_limit = q3+1.5*IQR 
lower_limit = q1-1.5*IQR
```

```python
df['ApplicantIncome'] = np.where(df['ApplicantIncome']>upper_limit,upper_limit,np.where(df['ApplicantIncome']<lower_limit,lower_limit,df['ApplicantIncome']))
```

```python
sns.boxplot(df.ApplicantIncome)
```

```python
sns.boxplot(df.CoapplicantIncome)
```

```python
q1 = df.CoapplicantIncome.quantile(0.25)
q3 = df.CoapplicantIncome.quantile(0.75)
IQR = q3-q1
upper_limit = q3+1.5*IQR 
lower_limit = q1-1.5*IQR
```

```python
df['CoapplicantIncome'] = np.where(df['CoapplicantIncome']>upper_limit,upper_limit,np.where(df['CoapplicantIncome']<lower_limit,lower_limit,df['CoapplicantIncome']))
```

```python
sns.boxplot(df.CoapplicantIncome)
```

```python
sns.boxplot(df.LoanAmount)
```

```python
q1 = df.LoanAmount.quantile(0.25)
q3 = df.LoanAmount.quantile(0.75)
IQR = q3-q1
upper_limit = q3+1.5*IQR 
lower_limit = q1-1.5*IQR
```

```python
df['LoanAmount'] = np.where(df['LoanAmount']>upper_limit,upper_limit,np.where(df['LoanAmount']<lower_limit,lower_limit,df['LoanAmount']))
```
